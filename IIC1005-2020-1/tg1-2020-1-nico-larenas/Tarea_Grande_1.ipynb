Parte 1


```
data = pd.read_csv('movies.csv', encoding="latin-1")
data
```

Pregunta 1

```
#https://www.it-swarm.dev/es/python/redondear-una-sola-columna-en-pandas/1049265738/
data1= data.copy()
data1.score = data.score.round()

#https://stackoverflow.com/questions/38933071/group-by-two-columns-and-count-the-occurrences-of-each-combination-in-pandas
ds = data1.groupby('score').size()
new_df = ds.to_frame(name = 'cantidad').reset_index()
new_df

alt.Chart(new_df).mark_bar(size=30).encode(
    x=alt.X('score', title='Puntajes'),
    y=alt.Y('cantidad', title='Cantidad de películas')
).properties(title="Gráfico 1: Distribución de puntajes")



```

Pregunta 2

```
top_10 = data.nlargest(10, "score")


#https://altair-viz.github.io/gallery/bar_chart_sorted.html
#https://stackoverflow.com/questions/43327125/pandas-max-value-in-column-and-subtract
alt.Chart(top_10).mark_bar(size=30).encode(

    x=alt.X('name', title='Nombre', sort='-y'),

    y=alt.Y('score', title='Puntajes'),

    color=alt.condition(

        alt.datum.score == top_10['score'].max() ,  

        alt.value('orange'),     

        alt.value('steelblue')
    )
).properties(title="Gráfico 2: Las mejores películas",width=alt.Step(40))


```

Pregunta 3

```
#http://www.datasciencemadesimple.com/drop-delete-rows-conditions-python-pandas/
new = data[data.budget != 0.0]
new

alt.Chart(new).mark_point().encode(
    x='budget',
    y='score',
    color='rating',
).properties(title="Gráfico 3: ¿Importa el presupuesto?")
```

Pregunta 4

```
#https://datascience.stackexchange.com/questions/33879/panda-grouping-by-month-with-transpose
data['month'] = pd.to_datetime(data['released']).dt.strftime('%m')
dr = data.groupby('month')['score'].mean()

new_df = dr.to_frame(name = 'promedio').reset_index()
new_df

alt.Chart(new_df).mark_bar(size=30).encode(
    x=alt.X('month', title='Mes'),
    y=alt.Y('promedio', title='Puntaje')
).properties(title="Gráfico 4: ¿Importa el mes?",width=alt.Step(40))

```
